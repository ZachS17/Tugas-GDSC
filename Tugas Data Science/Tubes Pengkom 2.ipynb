```python
import pandas as pd
import matplotlib.pyplot as plt


sn = "Canada by Citizenship" # Sebagai variabel untuk membaca sheet dari data yang diambil
df1 = pd.read_excel ("Canada.xlsx", sheet_name = sn) #variabel untuk membaca data dari file tersebut

df1korel=df1.corr()
df1korel

# Variabel untuk mengambil semua data dari data yang sudah dicari korelasinya
df2 = df1korel.iloc[:,:]

# Visualisasi dalam bentuk grafik bertipe bar chart
df2.plot(kind="bar", title="Korelasi antar atribut kuantitatif")
```

```python
# Karakteristik data pada atribut kuantitatif AREA dan REG

df4 = df1.iloc[:196,2:6]
df4.describe(include='all')
```

```python
# Karakteristik data pada atribut kuantitatif tahun 1980 sampai 2013

df2 = df1.iloc[:196,9:43]
df2.describe()
```

```python
# Membandingkan kategori

# Membandingkan jumlah negara setiap region (bar chart)
df2 = df1.groupby("Region").size()
df2[0:len(df2)-2].plot(kind="bar", x = "Region", title = "Jumlah Negara Setiap Region")
plt.ylabel("Jumlah Negara")
plt.ylim(2)
plt.legend(['Jumlah Negara'])
plt.show()
print("Penggunaan warna sama semua karena tidak ada kategori lagi di dalamnya dan tidak ada perbandingan antar kategori dan hanya meninjau jumlah.")
print("Insight dan informasi yang didapatkan yaitu dapat dilihat region mana yang jumlah negaranya paling banyak imigran ke Kanada, tanpa meninjau jumlah imigran setiap negaranya.")
print("Data ini didapatkan dari kolom region dan mendapatkan ukuran dari berapa kali region tertulis. Dibuat beberapa tambahan seperti lim agar batas bawah 2 dan interval 20 serta ditambah label dan legend.")

# Membandingkan jumlah negara berkembang dan maju
df2 = df1.groupby("Development region").size()
df2[0:len(df2)-2].plot(kind = "barh", title = "Area Berkembang dan Maju")
plt.ylabel("Negara Berkembang/Maju")
plt.legend(['Jumlah Negara'])
plt.show()
print("Penggunaan warna sama semua karena tidak ada kategori lagi di dalamnya dan tidak ada perbandingan antar kategori dan hanya meninjau jumlah.")
print("Insight dan informasi yang didapatkan yaitu dapat dilihat berapa dan perbandingan banyaknya negara berkembang dan negara maju yang imigrasi ke Kanada, tanpa melihat jumlah setiap negaranya.")
print("Data ini didapatkan dari kolom Development Region dan dilihat berapa negara yang memiliki kode 901 (negara maju) dan kode 902 (negara berkembang). Ditambahkan legend yaitu jumlah negara.")
```

```python
# Perubahan terhadap waktu

# Perubahan jumlah imigran Cina dari 1980-2013
df2 = df1.loc[df1["Major area"] == 935]
df2 = df1.iloc[37,9:43]
df2.plot(kind = "line", title = "Imigran Cina dari 1980-2013")
plt.legend(["Cina"])
plt.show()
print("Penggunaan warna hanya 1 karena tidak ada kategori lagi di dalamnya dan tidak ada perbandingan antar kategori dan hanya meninjau jumlah imigran pada sebuah negara.")
print("Insight dan informasi yang didapatkan yaitu dapat dilihat jumlah imigran pada sebuah negara pada rentang waktu yang panjang, kemudian dapat diolah misalnya apakah ada kejadian atau fenomena di negara tertentu yang menyebabkan perubahan?")
print("Data ini didapatkan dari mencari index dari data Cina dan kemudian diprint dari kolom 9 hingga 43. Ditambahkan legend yaitu Cina.")

# Perubahan jumlah total imigran setiap tahun dari 1980-2013
df3 = df1.iloc[197,9:43]
df3.plot(kind="line", title = "Total Imigran 1980-2013")
plt.legend(["Dunia"])
plt.show()
print("Penggunaan warna hanya 1 karena tidak ada kategori lagi di dalamnya dan tidak ada perbandingan antar kategori dan hanya meninjau jumlah imigran dunia.")
print("Insight dan informasi yang didapatkan yaitu dapat dilihat total jumlah imigran seluruh dunia, yang kemudian dapat diolah dalam skala global, apakah ada sebuah kejadian atau fenomena dan seberapa jauh perbedaannya?")
print("Data ini didapatkan dari baris paling bawah karena sudah mengandung total dunia. Ditambahkan legend yaitu Dunia.")

```

```python
# Hierarki dan hubungan keseluruhan bagian

# Hierarki region dalam negara berkembang dan maju
df2 = df1.groupby(["Development region","Region"])["Origin/Destination"].size().unstack()
df2.plot(kind = "bar", title = "Area Berkembang dan Maju", stacked = True)
plt.show()

print("Penggunaan warna berbeda semua karena menunjukkan setiap region yang berbeda agar bisa dibandingkan kontribusinya di negara berkembang dan maju.")
print("Insight dan informasi yang didapatkan yaitu dapat dilihat perbandingan jumlah negaranya pada setiap region dan dapat dilihat berapa diantaranya yang merupakan negara berkembang sehingga dapat diolah lebih lanjut seperti mengapa region dengan banyak negara maju tetap imigrasi ke Kanada.")
print("Data ini didapatkan dari pengelompokan apakah negara berkembang atau maju, kemudian dikelompokkan lagi berdasarkan regionnya dan dilakukan sebanyak negara yang ada dalam data. Legend tidak ditambahkan karena sudah ada karena merupakan stacked bar chart.")


# Hierarki jumlah imigran tahun 1980 dari Afrika Utara
df3 = df1.loc[df1["Region"] == 912]
df4 = df3.iloc[:,10]
df4.plot(kind="pie", title = "Hierarki Imigrant Tahun 1980 Afrika Utara")
plt.xlabel('Negara')
plt.ylabel("Imigran Afrika Utara")
plt.legend(df3.iloc[:,2])
plt.show()

print("Penggunaan warna berbeda semua karena menunjukkan setiap negara yang berbeda dari sebuah region (Afrika Utara) agar bisa dibandingkan jumlah imigran setiap negaranya terhadap keseluruhan imigran dari sebuah region.")
print("Insight dan informasi yang didapatkan yaitu dapat dilihat perbandingan jumlah imigran setiap negara dalam sebuah region dan jika diperluas jangka waktu dan jumlah regionnya bisa dapat diolah lebih lanjut seperti mengapa sebuah negara berkurang imigrannya padahal secara keseluruhan sebuah region banyak imigrannya.")
print("Data ini didapatkan pertama dari membuat sebuah subset informasi negara yang merupakan bagian dari region Afrika Utara, kemudian langsung dibuat dengan melihat jumlah imigran setiap negara pada tahun 1980. Legend ditambahkan untuk menunjukkan nama negaranya.")
```
